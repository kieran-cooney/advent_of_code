```python
def parse_input(filepath):
    with open(filepath, 'r') as f:
        out = [[*(l.strip().split('-'))] for l in f]
    return out
```

```python
test_inputs = [
    parse_input('test_input{}.txt'.format(i)) for i in [1,2,3]
]
```

```python
def find_neighbours(node, edges):
    out = (
        [n for n1, n in edges if n1 == node] +
        [n for n, n1 in edges if n1 == node]
    )

    return out
```

```python
def count_paths(edges, start_node='start', end_node='end'):
    count = 0

    def explore_cave(node, visited_small_nodes):
        nonlocal count

        if node == end_node:
            count += 1
            pass

        visited_small_nodes_c = visited_small_nodes.copy()
        if node.islower():
            small_nodes = visited_small_nodes + [node]
        else:
            small_nodes = visited_small_nodes
        
        neighbours = [
            n for n in find_neighbours(node, edges)
            if n not in small_nodes
        ]

        for n in neighbours:
            explore_cave(n, small_nodes)

    explore_cave(start_node, [])

    return count
```

```python
def count_paths_2(edges, start_node='start', end_node='end'):
    count = 0
    paths = []

    def explore_cave(node, visited_small_nodes, doubled, path):
        nonlocal count

        if node == end_node:
            count += 1
            paths.append(path)
            return

        visited_small_nodes_c = visited_small_nodes.copy()
        if node.islower():
            small_nodes = visited_small_nodes + [node]
        else:
            small_nodes = visited_small_nodes
        
        neighbours = [
            n for n in find_neighbours(node, edges)
            if n != start_node
        ]
        
        if doubled:
            neighbours = [n for n in neighbours if n not in small_nodes]

        will_double = [(n in small_nodes) or doubled for n in neighbours]

        for n, d in zip(neighbours, will_double):
            explore_cave(n, small_nodes, d, path + [n])

    explore_cave(start_node, [], False, [start_node])

    return count, paths
```
